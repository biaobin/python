```python
import sys
sys.path.append("/global/homes/b/bbli/app/python")

import bblplot as blt
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.loadtxt("./100E6_1degDogleg_TSCON_CSRON/fort.1027")

x  =  X[:,0]*1e3      #(mm)
y  =  X[:,2]*1e3      #(mm)
z  = -X[:,4]*1e3      #(mm)

px    = X[:,1]*0.511001  #(MeV)
py    = X[:,3]*0.511001  #(MeV)
dgam  = -X[:,5]*0.511001 #(MeV)
```

```python
plt.figure
blt.zdgamplot(z,dgam)
blt.plot2d(z, x, xlabel='z (mm)', ylabel='x (mm)', filename='zx.png')
blt.plot2d(z, y, xlabel='z (mm)', ylabel='y (mm)', filename='zy.png')
blt.plot2d(x, dgam, xlabel='x (mm)', ylabel='$\Delta$E (MeV)', filename='xdgam.png')
blt.plot2d(px, dgam, xlabel='px (MeV/c)', ylabel='$\Delta$E (MeV)', filename='pxdgam.png' )
```
